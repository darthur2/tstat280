#### Part 1

Using the food delivery dataset, complete the following tasks:

1. **Create a new column** called `total_time` that equals the sum of `food_preparation_time` and `delivery_time`.

2. **Filter the dataset** to only include orders placed on the **Weekend**.

3. **Sort the filtered results** by `cost_of_the_order` in **descending order**.

4. Display the **top 5 most expensive weekend orders**.

```python
import pandas as pd

df = pd.read_csv("nyc_food.csv")

df["total_time"] = df["food_preparation_time"] + df["delivery_time"]

df[df["day_of_the_week"] == "Weekend"].sort_values(by="cost_of_the_order",
                                                  ascending=False).head()
```

#### Part 2

Using the food delivery dataset, complete the following tasks:

1. Create a new column called `is_cheap_and_fast` using **`apply()` and a lambda function**.
   - An order should be labeled **1** if the `cost_of_the_order` is **less than 25** **and** the `total_time` is **less than 40**.
   - Otherwise label it **0**.

2. Group the data by `cuisine_type`.

3. For each cuisine type, calculate:
   - the **average cost of the order**
   - the **total number of cheap and fast orders**

```python
df["is_cheap_and_fast"] = df.apply(lambda row:
                                   1 if (row["cost_of_the_order"] < 25 and row["total_time"] < 40) else 0,
                                   axis = 1)

df.groupby("cuisine_type").agg({"cost_of_the_order": "mean",
                               "is_cheap_and_fast": "sum"})
```
