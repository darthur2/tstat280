## What is Pandas?

**Pandas** is a Python library for working with structured data.

It provides tools to:
- Read data from files and databases  
- Explore and summarize datasets  
- Clean and transform variables  
- Perform data analysis  

The main data structure in Pandas is the **DataFrame**, which is similar to a spreadsheet or database table.

## Reading Files

## Reading CSV Files

Reading CSV files means importing data stored in a CSV (Comma-Separated Values) file into Python. The `pd.read_csv` function reads the file and returns a **DataFrame**, allowing the data to be explored and analyzed in Python.

```python
import pandas as pd

df = pd.read_csv("../data/social_media.csv")

type(df)
```

## Querying SQL Databases

Querying a SQL database means retrieving data from a database using SQL and loading it into Pandas. The `pd.read_sql` function runs a SQL query and returns the result as a **DataFrame**, allowing the data to be analyzed in Python.

```python
import sqlite3

conn = sqlite3.connect("../data/movies.sqlite")

df_sql = pd.read_sql("""
    SELECT * FROM movies
    LIMIT 10
""", conn)

type(df_sql)
```

## Filtering and Indexing

## Selecting a Column

Selecting a column means extracting a single variable from a DataFrame. A column can be selected using `df["column"]`, which returns the selected column as a **Series**, allowing you to analyze or transform that variable.

```python
df["Country"].head()
```

## Selecting Multiple Columns

Selecting multiple columns allows you to extract more than one variable from a DataFrame at the same time. This is commonly done using `df[["column1"], ["column2"]]`.

```python
df[["Age", "Academic_Level"]].head()
```

## Selecting by Row's Name

Selecting by a row's name allows you to access a specific row using its label (index name). This can be done with `df.loc[row_index]`

```python
df.loc[5:9]
```

## Selecting Rows and Columns by Name

You can select a specific row and column using their labels with `.loc`. This can be done by `df.loc["row_name", "column_name"]`

```python
df.loc[5:9, ["Age", "Most_Used_Platform"]]
```

## Selecting a Row by Index

You can select a row based on its numeric position in the DataFrame using `.iloc`. This can be done with `df.iloc[index]`.

This returns the row located at the specified index position.

```python
df.iloc[5:10]
```

## Selecting a Column by Index

You can select a column based on its numeric position using `.iloc`. This can be done with `df.iloc[:, index]`.

This returns the column located at the specified index position.

```python
df.iloc[:, [1, 6]].head()
```

## Selecting Rows and Columns by Index

You can select specific rows and columns using their numeric positions with `.iloc`. This can be done with `df.iloc[row_index, column_index]`.

This returns the value located at the specified row and column positions.

```python
df.iloc[5:10, [1, 6]].head()
```

## Filtering Using Conditions

You can filter rows in a DataFrame by applying a condition to a column. This can be done with `df.loc[filter_condition, "column"]`.

This returns values from the specified column for rows that satisfy the condition.

```python
df.loc[df["Age"] < 20, "Most_Used_Platform"]
```

## Sorting

## Sorting by Column

You can sort the rows of a DataFrame based on the values in a column using `.sort_values()`. This can be done with `df.sort_values(by="column")`.

This returns the DataFrame with rows ordered according to the values in the specified column.

```python
df.sort_values(by="Avg_Daily_Usage_Hours").head()
```

## Sorting (Ascending or Descending)

You can control the sort direction using the `ascending` argument in `.sort_values()`. This can be done with `df.sort_values(by="column", ascending=True)`.

Set `ascending=True` for increasing order or `ascending=False` for decreasing order.

```python
df.sort_values(by="Avg_Daily_Usage_Hours", ascending=False).head()
```

## Sorting by Multiple Columns

You can sort a DataFrame using more than one column. This can be done with `df.sort_values(by=["column1", "column2"])`.

Rows are first sorted by the first column, and ties are then sorted using the next column.

```python
df.sort_values(by=["Age", "Gender"]).head()
```

## Manipulating Columns

## Creating Columns

You can create a new column by assigning values to a new column name. This can be done with `df["new_column"] = new_column_values`.

This adds the new column to the DataFrame.

```python
df["Sleep_Usage_Ratio"] = (df["Sleep_Hours_Per_Night"] / 
                           df["Avg_Daily_Usage_Hours"])

df["Sleep_Usage_Ratio"].head()
```

## Renaming Columns

You can rename columns in a DataFrame using `.rename()`. This can be done with `df.rename(columns={"old_name": "new_name"})`.

This changes the column name while leaving the rest of the DataFrame unchanged.

```python
df.rename(columns={"Avg_Daily_Usage_Hours": "Usage",
                   "Sleep_Hours_Per_Night": "Sleep"}, inplace=True)

df.head()
```

## Dropping Columns

You can remove a column from a DataFrame using `.drop()`. This can be done with `df.drop(columns="column")`.

This returns a DataFrame with the specified column removed.

```python
df.drop(columns=["Student_ID", "Sleep_Usage_Ratio"], inplace=True)

df.head()
```

## Lambda Functions

## What is a Lambda Function?

A **lambda function** is a small, anonymous function that can be defined in a single line.

In Python, lambda functions are often used for quick operations that do not require defining a full function. They are commonly used with Pandas functions like `map()` and `apply()` to transform data.

```python
import numpy as np

numbers = np.array([0, 1, 2, 3, 4])

my_funct = lambda x: x**2

my_funct(numbers)
```

## What is map?

`map()` is a Pandas method used to transform the values in a single column.

It applies a function or mapping to each value in a Series. This can be done with `df["column"].map(function)` or `df["column"].map(dictionary)`.

```python
# Example of using map with a dictionary
academic_level_dict = {"High School": "HS", 
                       "Undergraduate": "U", 
                       "Graduate": "G"}

df["Academic_Level_Short"] = df["Academic_Level"].map(academic_level_dict)

# Example of using map with a function
df["Gender_Int"] = df["Gender"].map(lambda x: 1 if x == "Female" else 0)
```

## What is map? (Continued)

```python
df[["Academic_Level", "Academic_Level_Short", "Gender", "Gender_Int"]].head()
```

## What is apply?

`apply()` is a Pandas method used to apply a function to data in a DataFrame or Series.

It is often used with lambda functions to perform operations across rows or columns. This can be done with `df.apply(function)` or `df["column"].apply(function)`.

```python
df["Sleep_Usage_Ratio"] = df.apply(lambda row: 
                                   row["Sleep"]/row["Usage"], axis = 1)

df[["Sleep", "Usage", "Sleep_Usage_Ratio"]].head()
```

## Grouping and Summarizing

## Getting Non-Null Counts

You can count the number of non-missing values in a column using `count()`. This can be done with `df["column"].count()`.

This returns the number of observations that are not null in the selected column.

```python
print(df["Gender"].count())

print(df[["Gender", "Academic_Level_Short"]].count())
```

## Getting Category Counts

You can count how many times each value appears in a column using `value_counts()`. This can be done with `df["column"].value_counts()`.

This is commonly used to summarize **categorical variables**.

```python
print(df["Academic_Level"].value_counts())

df[["Relationship_Status", "Gender"]].value_counts().head()
```

## Getting Bivariate Categorical Counts

You can view counts for combinations of two categorical variables using `pd.crosstab()`. This can be done with `pd.crosstab(df["column1"], df["column2"])`.

This produces a table showing how frequently each pair of categories occurs.

```python
pd.crosstab(df["Gender"], df["Relationship_Status"])
```

## Seeing Correlations

You can compute correlations between numeric variables using `df.corr()`.

This returns a correlation matrix showing the strength and direction of relationships between quantitative columns.

```python
numeric_columns = ["Usage", "Sleep", "Conflicts_Over_Social_Media", "Addicted_Score"]

df[numeric_columns].corr()
```

## Grouping and Summary Functions

You can group data by a variable and compute summary statistics using `groupby()`. This can be done with `df.groupby("group_column")["target_column"].function()`.

This allows you to calculate statistics such as the mean, sum, or count for each group.

```python
df.groupby("Gender")["Age"].mean()
```

## Group By and Aggregate

You can compute multiple summary statistics for multiple columns using `.groupby()` with `.agg()`. This can be done with `df.groupby("group_column")[["target_column1", "target_column2"]].agg(["function1", "function2"])`.

This returns grouped summaries for the selected columns using the specified functions.

```python
df.groupby("Academic_Level")[["Usage", "Sleep"]].agg(["mean", "median"])
```

## Group By and Aggregate

An alternative way to compute multiple summary statistics is using `.groupby()` and `.agg()` is to provide a dictionary specifying which summaries to compute for different columns. This is done with `df.groupby("group_column").agg({"target_column1": ["function1", "function2"], "target_column2": ["function1", "function2"]})`.

```python
df.groupby("Academic_Level").agg({"Usage": ["mean", "median"], "Sleep": ["min", "max"]})
```

## Merging DataFrames

You can combine two DataFrames based on matching keys using `merge()`. This can be done with `df.merge(right_df, how="join_type", left_on="left_key", right_on="right_key")`. If the key names are the same for both DataFrames, you dan do `df.merge(right_df, how="merge_type", on="key")`. Inner joins are performed by default.

This joins the rows from both DataFrames where the specified keys match.

```python
df_movies = pd.read_sql("""
    SELECT * FROM movies
""", conn)

df_directors = pd.read_sql("""
    SELECT * FROM directors
""", conn)

df_md = df_directors.merge(df_movies, how="left", left_on="id", right_on="director_id")

df_md.loc[0:1, ["name", "title"]]
```

## Converting Variable to DateTime

You can convert a column to a datetime type using `pd.to_datetime()`. This can be done with `pd.to_datetime(df["date_column"])`.

This allows you to work with dates and times more easily in Pandas.

```python
print(type(df_movies["release_date"][0]))

df_movies["release_date"] = pd.to_datetime(df_movies["release_date"])

print(type(df_movies["release_date"][0]))
```
