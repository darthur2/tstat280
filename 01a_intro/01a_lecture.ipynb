## The Data Science Workflow

## Learning Objectives

1. Identify and describe the different components of the data science workflow
2. Distinguish between supervised and unsupervised learning problems
3. Distinguish between regression and classification problems
4. Identify and distinguish between different roles in data science

## What is Data Science?

**Data science** is the field of extracting knowledge and insights from data using statistics, computing, and domain expertise.

**Examples of core activities:**

* Collect and manage data
* Clean and prepare data
* Explore and visualize data
* Build statistical or machine learning models
* Communicate insights to support decisions


## Why Data Science?

Organizations use **data science** to turn data into better decisions.

**Why it matters:**
- Make **data-driven decisions**
- **Discover patterns** in large datasets
- **Predict future outcomes**
- **Improve efficiency** and reduce costs
- Gain **competitive advantage**

**Examples:**
- Product recommendations  
- Fraud detection  
- Sales forecasting  
- Operations optimization

## The Data Science Workflow

There is **no single authoritative definition** of the data science workflow.  
Different organizations use different processes, but most workflows include similar stages.

**Common steps:**
1. Define the **problem or question**
2. **Collect and acquire data**
3. **Clean and prepare** the data
4. **Explore and analyze** the data
5. **Build models** or apply statistical methods
6. **Evaluate results**
7. **Communicate results** and deploy models

## Problem Definition

The **first step** in the data science workflow is clearly defining the problem.

**Key questions:**
- What decision needs to be made?
- What question are we trying to answer?
- Who are the stakeholders?
- What does **success** look like?

**Why it matters:**
- Aligns the project with **business goals**
- Determines what **data** is needed
- Guides the **analysis and modeling approach**

## Supervised Learning

**Supervised learning** uses data with **known outcomes (labels)** to train models.

**Goal:**
- Learn a relationship between **inputs (features)** and **outputs (labels)**
- Use that relationship to **predict outcomes for new data**

**Examples:**
- **Regression:** Predicting a student's **final exam score** based on study hours and attendance  
- **Classification:** Classifying **emails as spam or not spam**

## Regression vs Classification

**Regression**
- Used when the output is a **continuous value**
- The model predicts a **number**

**Classification**
- Used when the output is a **category or label**
- The model predicts a **class**

**Examples:**
- **Regression:** Predicting a house **price** based on size and location  
- **Classification:** Classifying a tumor as **benign or malignant**

## Unsupervised Learning

**Unsupervised learning** works with data that **does not have labels**.

**Goal:**
- Discover **patterns, structure, or groupings** in the data

**Examples:**
- **Customer segmentation** based on purchasing behavior  
- **Grouping news articles** by similar topics

## Collect and Acquire Data

Once the problem is defined, the next step is gathering the data needed for analysis.

**Common data sources:**
- Databases and data warehouses  
- APIs and web services  
- Surveys and experiments  
- Sensors, logs, or transaction records  
- Public or open datasets

**Key considerations:**
- Is the data **relevant** to the problem?
- Is the data **complete and reliable**?
- Do we have permission to **access and use** the data?

## Clean and Prepare Data

Raw data is often **messy or incomplete**, so it must be cleaned and prepared before analysis.

**Common tasks:**
- Handle **missing values**
- Fix **errors or inconsistencies**
- Remove **duplicates**
- Convert data into usable **formats**
- Create new **features or variables**

**Why it matters:**
- Improves **data quality**
- Makes data **ready for analysis and modeling**

## Explore and Analyze Data

Before building models, data scientists explore the data to understand its structure and patterns.

**Common activities:**
- Calculate summary statistics
- Visualize data with charts and plots
- Identify patterns, trends, or anomalies
- Examine relationships between variables

**Goals:**
- Develop intuition about the dataset
- Identify potential issues or outliers
- Inform the choice of modeling approaches

## Build Models

In this stage, data scientists apply statistical or machine learning methods to learn patterns from the data.

**Common activities:**
- Select an appropriate modeling approach
- Train models using historical data
- Tune model parameters
- Compare multiple models

**Examples:**
- Regression for predicting numeric values
- Classification for predicting categories
- Clustering to find natural groupings in data

## Evaluate Models and Results

After building models, the next step is to determine how well they perform.

**Common activities:**
- Test models on **unseen data**
- Calculate **performance metrics**
- Compare results across multiple models
- Check for **overfitting or bias**

**Examples of metrics:**
- Accuracy
- Mean squared error

**Goal:**
- Ensure the model provides **reliable and useful results**.

## Communicate Results and Deploy Models

The final step is turning analysis into **actionable outcomes**.

**Communicating results:**
- Present findings with **visualizations and reports**
- Explain insights to **stakeholders**
- Provide **recommendations** for decisions

**Model deployment:**
- Integrate models into **applications or systems**
- Automate predictions or decision support
- Monitor model **performance over time**

**Goal:**
- Ensure insights and models create **real-world impact**.

## Roles in Data Science

Data science projects often involve **multiple roles working together**.

**Common roles:**
- **Stakeholders** – define the problem, set goals, and use results to make decisions  
- **Domain Experts (SMEs)** – provide subject-matter knowledge and help interpret results in context  
- **Data Engineers** – collect, store, and manage data  
- **Data Scientists / Statisticians** – analyze data and build models  
- **Data Analysts** – explore data and communicate insights  
- **Machine Learning Engineers** – deploy and maintain models in production  

> **Note:**  
> - **Stakeholder** = “What problem should we solve?”  
> - **Domain Expert** = “Do these results make sense in the real world?”

## Stakeholders

**Stakeholders** define the problem, set goals, and ensure the work aligns with organizational needs.

**Responsibilities:**
- Define the **business problem or question**
- Set **goals and success criteria**
- Prioritize what problems are most important to solve
- Use insights and model outputs to **inform decisions**

**Examples:**
- Marketing managers  
- Product managers  
- Business executives  
- Operations managers  

## Domain Experts (SMEs)

**Domain Experts (Subject Matter Experts)** provide specialized knowledge about the field and help ensure that results are meaningful and realistic.

**Responsibilities:**
- Provide **context and domain knowledge**
- Help identify which variables and features are important
- Interpret results and determine if they **make sense in the real world**
- Suggest improvements based on industry knowledge

**Examples:**
- Healthcare professionals (doctors, nurses)  
- Financial analysts  
- Engineers  
- Retail or marketing specialists  

## Data Engineers

**Data engineers** build and maintain the systems that collect, store, and process data.

**Responsibilities:**
- Build **data pipelines** to collect and move data
- Manage **databases and data warehouses**
- Ensure data is **reliable, accessible, and scalable**
- Prepare data for **analysis and modeling**

**Tools often used:**
- SQL
- Distributed data systems
- Data pipelines and ETL workflows

## Data Scientists / Statisticians

**Data scientists and statisticians** analyze data to extract insights and build predictive models.

**Responsibilities:**
- Explore and analyze datasets
- Apply **statistical methods** and **machine learning**
- Build and evaluate predictive models
- Interpret results and generate insights

**Common tools:**
- Python or R
- Statistical modeling
- Machine learning methods

## Data Analysts

**Data analysts** focus on exploring data and communicating insights that support decision-making.

**Responsibilities:**
- Query and analyze data
- Create **reports and dashboards**
- Visualize data to highlight trends
- Communicate insights to stakeholders

**Common tools:**
- SQL
- Excel or spreadsheets
- Data visualization tools (e.g., Tableau, Power BI)

## Machine Learning Engineers

**Machine learning engineers** focus on deploying and maintaining models in real-world systems.

**Responsibilities**
- Deploy models into **applications or production systems**
- Build **APIs and services** for model predictions
- Monitor **model performance over time**
- Retrain and update models as new data arrives

**Common tools**
- Python
- Cloud platforms
- Model serving and monitoring tools
