### Exercises

#### Exercise 1

**Goal:** Practice making a change and committing it using GitHub Desktop

Do the following:

1. Open your project repository in GitHub Desktop  
2. In your project folder, create a new file called `hello.py`  
3. Add the following code to the file:
```python
print("Hello, world!")
```
4. Save the file  
5. Add a commit message
6. Push the changes to your GitHub repo

#### Exercise 3

**Goal:** Create a new environment with different versions than the course environment

Do the following:

1. Create a new environment called `test-env` with a **different version of Python** than the one used in the course (`3.10`):
```bash
conda create -n test-env python=3.11
```
2. Activate the environment:
```bash
conda activate test-env
```
3. Install **one library from the course list** (e.g., numpy, pandas, matplotlib, etc.) but choose a **different version** than what is specified in the `tstat280` environment. For example:
```bash
conda install numpy=1.25
```
4. Wait for the installation to complete
5. Deactivate the test environment
```bash
conda deactivate
```

#### Exercise 4

**Goal:** Verify that environments can have different Python and package versions

Do the following:

1. Activate the original course environment:
```bash
conda activate tstat280
```
2. Check the Python version:
```bash
python --version
```
3. Check the version of the library you installed in Exercise 1:
```bash
conda list numpy
```
4. Deactivate the environment and activate your test environment:
```bash
conda deactivate
conda activate test-env
```
5. Check the Python version again:
```bash
python --version
```
6. Check the version of the same library again:
```
conda list numpy
```
7. Deactivate the test environment
```bash
conda deactivate
```
7. Remove the test environment when finished:
```bash
conda remove -n test-env --all
```
