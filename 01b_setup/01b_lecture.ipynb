## Version Control

## Git

**Git** is a **version control system** used to track changes in code and collaborate with others.

**What Git does:**
- Tracks changes to files over time
- Allows you to **save versions** of your work
- Helps teams **collaborate on the same project**

**Key concepts:**
- **Repository (repo):** a project folder tracked by Git
- **Commit:** a saved snapshot of changes
- **Push:** upload changes to a remote repository
- **Pull:** download updates from a remote repository

## Why Use Version Control in Data Science?

Version control tools like Git help you manage and organize your work.

- **Track changes** – Save versions and go back if something breaks  
- **Experiment safely** – Try new ideas without losing past work  
- **Collaborate** – Work with others without overwriting changes  
- **Reproduce results** – Keep a record of how results were created  

**In short:** Version control keeps your work organized, safe, and reproducible.

## GitHub

**GitHub** is an online platform for **hosting and collaborating on Git repositories**.

**What GitHub provides:**
- A place to **store and share repositories**
- Tools for **collaboration and code review**
- A way to **track issues and project progress**

**Common actions:**
- **Create repositories** to store projects
- **Fork repositories** to copy someone else's project
- **Clone repositories** to work on them locally
- **Push changes** from your computer to GitHub

## What is a Repository?

A **repository (repo)** is a folder that stores a project and its history using Git.

**What a repository contains**
- Project **files and code**
- **Version history** of changes
- Documentation (often a **README** file)

**Why repositories are useful**
- Track changes over time
- Collaborate with others
- Manage and organize projects

## Creating a GitHub Repository

A **repository** is where your project files and history are stored on GitHub.

**Why create a repository**
- Store and organize **project files**
- **Track changes** with Git
- **Share or collaborate** with others

**How to create a repository**
1. Click the **+** icon on GitHub  
2. Select **New repository**  
3. Enter a **repository name**  
4. Choose **Public** or **Private**  
5. Click **Create repository**


## Creating a Repository from a Template

A **template repository** lets you create a **new project that starts with existing files**.

**Why use a template**
- Start with **pre-made project files**
- Keep your work **separate from the original**
- Ensure everyone begins with the **same materials**

**How to create a repository from a template**
1. Go to the **template repository**
2. Click **Use this template**
3. Select **Create a new repository**
4. Enter a **repository name**
5. Click **Create repository**


## Adding Someone With View Access

You can allow someone to **view your repository without editing it**.

**Why give view access**
- Let an instructor **review your work**
- Share your project **without allowing edits**

**How to add view access**
1. Go to your **repository**
2. Click **Settings**
3. Select **Collaborators and teams**
4. Click **Add people**
5. Enter the **GitHub username**
6. Set role to **Read**


## GitHub Desktop

**GitHub Desktop** is a graphical application that makes it easier to work with **Git and GitHub**.

**What it allows you to do**

## Cloning a Repository (GitHub Desktop)

**Cloning** creates a **local copy of a repository** on your computer.

**Steps**
1. Open **GitHub Desktop**
2. Click **File → Clone Repository**
3. Select the repository from the **GitHub.com** tab
4. Choose the **local folder** where it will be saved
5. Click **Clone**

You can now **edit the project files locally** on your computer.

## Commit and Push

**Commit**
- A **commit** saves a snapshot of your changes in the repository.
- Each commit includes a **message** describing what changed.

**Push**
- **Push** sends your committed changes from your computer to **GitHub**.

**Typical workflow**
1. Make changes to files
2. **Commit** the changes in GitHub Desktop
3. **Push** the commits to GitHub

## Creating the Course Environment

**Navigate to your course folder**
- Use `cd` to move into the repository you cloned

**Create the environment**
- Run:
  - `conda env create -f environment.yml`
- This installs Python and all required packages

**Activate the environment**
- Run:
  - `conda activate tstat280`
- Your terminal should now show `tstat280`

## What Happens When You Create an Environment?

When you create a conda environment:

- **A separate Python installation is created** just for this project  
- **Specified packages are installed** (from `environment.yml`)  
- **Everything is isolated** from your other projects and system Python  

When you activate the environment:
- Your terminal switches to use this specific Python and its packages  

**In short:** You’re creating a self-contained workspace with its own Python and libraries.

## Managing the Environment

**Activate the environment**
- Run:
  - `conda activate tstat280`
- Use this every time before working on the course

**Deactivate the environment**
- Run:
  - `conda deactivate`
- Returns you to your normal system environment

**Reinstall (start over if needed)**
- If something breaks, run:
  - `conda remove -n tstat280 --all`
  - `conda env create -f environment.yml`

**Key idea**
- Environments can be safely deleted and recreated at any time
