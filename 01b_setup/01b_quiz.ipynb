#### Part 3: Running a Code Cell to Produce Output

Run the following code block below to output the name of the environment, the Python version, and the version of `numpy`.

````python
import sys
import numpy as np
import os

env_name = os.environ.get("CONDA_DEFAULT_ENV", "Unknown")

print(f"Environment: {env_name}")
print(f"Python version: {sys.version.split()[0]}")
print(f"Numpy version: {np.__version__}")
````
