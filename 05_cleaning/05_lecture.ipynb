## Checking for Duplicates

Early on we want to check for duplicates and remove them. If an observation is duplicated many times and we include all copies when modeling, this can bias or skew our analysis. We can check for duplicates using the `duplicated` method on the entire DataFrame or a column we expect to have unique values.

```python
duplicate_titles = df[df["product_title"].duplicated()]["product_title"]

df[df["product_title"] == duplicate_titles.iloc[0]]
```

## Dropping Duplicates

If you decide to drop observations because they appear like duplicates, then you can use the `drop_duplicates` method. For example, `df.drop_duplicates()` which drops rows that have the same values for every column or `df.drop_duplicates(subset="column")` which drops rwos that have the same values for a specific column.

```python
df.drop_duplicates(subset="product_title", inplace=True)

df
```

## Treating Missing Values

There are many ways to deal with missing values, each with its pros and cons. Here is just a small list of simple methods:

- Remove any rows where values are missing for any variables (**listwise deletion**)
- Remove variables with a lot of missing data (**variable deletion**)
- Remove only those rows where values are missing for specific variable (**case deletion**)
- Replace missing values with a measure of central tendency (e.g., the mean, **mean imputation**)

## Treating Missing Values: Pros and Cons

| Method           |Pros                                                |Cons                        |
|------------------|----------------------------------------------------|----------------------------|
|**Listwise Deletion**|Easy, valid if data are missing completey at random |You lose a lot of data, leads to bias if data aren't missing completely at random |
|**Variable Deletion**|Easy, you don't lose any observations, useful if that variable isn't important|Problematic if that variable would have been useful for modeling|
|**Case Deletion**|Easy, valid if data missing completely at random, you don't lose as much data|You still throw away some data, can still lead to bias if there are patterns to missingness|
|**Mean Imputation**|Still Easy, valid if data are missing completely at random, you don't lose any data|Can lead to serious bias if a lot of data are missing or data aren't missing completely at random|

## Listwise Deletion

If we wanted to delete any records (rows) with any variables with missing values, we could use `df.dropna()`.

```python
df_deleted = df.dropna()

df_deleted.info()
```

## Variable Deletion

If we want to get rid of certain columns, we can use `df.drop(labels=["column1", "column2"], axis=1)`.

```python
df.drop(labels=["Unnamed: 0", "next_month_bought", 
                "product_title", "product_details_dict",
                "technical_details_dict", "about"], axis=1, inplace=True)

df.info()
```

## Case Deletion

Whenever there are variables for which it might be difficult to impute values and there isn't a lot of missing data, we can remove the rows that have missing observations just for those values. This can be done using `df.dropna(subset=["column1", "column2"]))`

```python
df.dropna(subset=["model_name", "cpu_model", "os", "special_features"], inplace=True)

df.info()
```

## Mean (Median, Mode) Imputation

We can impute values for a variable using some measure of central tendency. For numeric variables, we often use the mean or the median and for categorical variables we may use the mode.

For example, to impute values using the mean or median, we can do `df["column"].fillna(df["column"].mean())` or `df["column"].fillna(df["column"].median())`.

```python
df["rating"] = df["rating"].fillna(df["rating"].mean())

df["rating"].describe()
```

## Removing Unwanted Text/Characters

When you want to remove unwanted text or characters from a column, an easy approach is to use the `replace` method. For example, you can do `df["column"].str.replace("string pattern", "replace pattern")`

```python
df["total_rating"].str.replace("(", "")
```

## Method Chaining

When we want to perform a sequence of operations, we can use **method chaining**. **Method chaining** is when multiple methods are called sequentially on a single object. For example, we can do `df["column"].method1().method2()`. You can span multiple lines by starting with `(` and ending with `)`.

```python
(
    df["total_rating"].str.replace("(", "")
    .str.replace(")", "")
    .str.replace(",", "")
)
```

## Data Type Conversion

Even after unwanted text or characters have been removed from a column, the column type will still be the same. If we want to treat the column as numeric, then we have to convert the data type. This can be done using `df["column"].astype("type")`. If a numeric variable has missing values, we will want to convert to the `float` data type.

```python
(
    df["total_rating"].str.replace("(", "")
    .str.replace(")", "")
    .str.replace(",", "")
    .astype("float")
)
```

## Clean, Convert, and Impute via Method Chaining

A common pipeline for a column is to clean, convert, and then impute. This can be done one step at a time or we can combine method chaining and the `pipe` method. The `pipe` method lets us perform an action on an entire column at once. For example, we can do `df["column"].method1().method2().pipe(function)` to use a function on an entire column.

```python
(
    df["total_rating"].str.replace("(", "")
    .str.replace(")", "")
    .str.replace(",", "")
    .astype("float")
    .pipe(lambda x: x.fillna(x.mean()))
)
```

## Standardizing Scales (Part I)

Different sources may report information using different metrics. In this case, it will be necessary to standardize the scales. First, we might want to determine the unit of measurement for each item. One way to do this is using `df["column"].str.contains("pattern")`.

```python
df["screen_size"].str.contains("Centimetres")
```

## Standardizing Scales (Part II)

Once we know what the the unit of measurement is, we can determine the multiplier for each value.

```python
CM_TO_IN = 0.393701

screen_multiplier = (
    df["screen_size"].str.contains("Centimetres").
    map(lambda boolean: CM_TO_IN if boolean else 1)
)

df["screen_size"] = (
    df["screen_size"].str.replace(" Centimetres", "")
    .str.replace(" Inches", "")
    .astype("float")*screen_multiplier
)

df["screen_size"].describe()
```

## Checking for Outliers (IQR Method)

It's important to understand if there are any unusual values. One way to do this is using the IQR method. To do this you will need to use the `quantile` method. For example `df["column"].quantile()`.

```python
df["price"] = df["price"].str.replace(",", "").astype("float")

Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3-Q1

df[(df["price"] < (Q1 - 1.5*IQR)) | (df["price"] > (Q3 + 1.5*IQR))].head()
```

## Replacing Text with Regular Expressions (Advanced; Optional)

While you can get by cleaning text without these, regular expressions (**regex**) can be very useful and can simplify many data cleaning tasks. For example, we can provide multiple characters at once when using `df["column"].str.replace(r"([put characters here])", "replace")`

```python
df["mrp"].str.replace(r"([₹,])", "", regex=True)
```

## Extracting Text with Regular Expressions (Advanced; Optional)

We can also use regular expressions to extract certain patterns. This can be done using `df["column"].str.extract(r"pattern")`. In the patterns you can use character classes `[0-9]` to match digits and `[A-Za-z]` to match letters. You can also use `+` to match one or more characters.

```python
df["hard_disk_size"].str.extract(r"([A-Za-z]+)")
```

## Going Further with Regular Expressions (Advanced; Optional)

Regular expression provide a flexible and powerful way to match any pattern you want. The pattern below has two capture groups, each in parenthesis `()`. The `|` is used as logical OR. The `?` says that the pattern is optional. Finally, the `\s` matches a white space.

```python
import re

df["colour"].str.extract(r"""(Luna|Steel|Mica|Iron|Mecha|
                              Natural|Graphite|Glacier|Platinum|
                              Mixed|Cloud|Off|Eclipse|Quiet|Performance|
                              Cool|Shadow|Carbon|Core|SpaceTurbo|
                              Ice|Eclipse|PLATINUM|Jaeger|Arctic|
                              Cosmos|Scandinavian|Ash|Stellar|
                              Translucent)?\s?(Grey|Gray|Black|
                              Silver|Blue|Aura|silver|blue|Green|
                              Aqua|GREY|White|Platinum)?""", flags=re.VERBOSE)
```
