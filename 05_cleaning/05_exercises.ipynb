#### Exercise 3

Use mean imputation to replace missing values in the Hits column.

```python
df["Hits"].str.contains("K")
```

```python
# YOUR CODE HERE

K_TO_1000 = 1000

hits_multiplier = (
    df["Hits"].str.contains("K").
    map(lambda x: K_TO_1000 if x else 1)
)

df["Hits"] = (
    df["Hits"].str.replace("K", "")
    .astype("float")*hits_multiplier
)

df["Hits"].describe()
```
