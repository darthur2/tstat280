# Databases and SQL

## Databases

A **database** is an organized collection of data stored electronically so it can be easily accessed, managed, and updated.

In many databases, data is stored in tables made up of rows and columns. Each row represents a single record, and each column represents an attribute of that record.

```python
# Importing sqlite3 library and pprint (pretty print) function
import sqlite3
from pprint import pprint

# Creating a connectiong to the database
conn = sqlite3.connect("movies.sqlite")

# The cursor will allow us to query the database
cur = conn.cursor()
```

# Selecting Columns and Rows

## SELECT, FROM, and LIMIT

The **SELECT** statement specifies which columns should be returned in the query results.

The **FROM** clause specifies the table that the data should be retrieved from.

The **LIMIT** clause restricts the number of rows returned by the query.

## SELECT, FROM, and LIMIT

```python
# Selecting all columns from the movies database
# Limits to the first 10 records (rows)

# Executes query and stores result
res = cur.execute("""
    SELECT * FROM movies
    LIMIT 1
""")

# Gets records
res.fetchall()
```

## Keys

A **primary key** is a column (or set of columns) that uniquely identifies each row in a table. No two rows can have the same primary key value, and the value cannot be NULL.

A **foreign key** is a column in one table that references the primary key in another table. Foreign keys create relationships between tables and help maintain consistent and connected data.

```python
movies = cur.execute("""
    SELECT * FROM movies
    LIMIT 0
""")

# List comprehension for seeing movie columns
print([col[0] for col in movies.description])
```

## Selecting Specific Columns

```python
res = cur.execute("""
    SELECT id, title, vote_average FROM movies
    LIMIT 5
""")

res.fetchall()
```

## DISTINCT

The **DISTINCT** keyword is used in a query to return only unique values from a column or set of columns.

When DISTINCT is used, duplicate rows in the result are removed so that each value appears only once.

```python
res = cur.execute("""
    SELECT DISTINCT director_id FROM movies
    LIMIT 5
""")

res.fetchall()
```

# Aggregate Functions

## Aggregate Functions

An **aggregate function** performs a calculation on multiple rows and returns a single summarized value.

Common aggregate functions include COUNT, SUM, AVG, MIN, and MAX, which are used to summarize data in a table.

## COUNT

The **COUNT** function returns the number of rows that match a query.

It is commonly used to determine how many records exist in a table or how many rows satisfy a particular condition.

```python
# Gets all movies
res = cur.execute("""
    SELECT COUNT(*), COUNT(DISTINCT director_id) FROM movies
""")

print(res.fetchall())
```

## AVG and SUM

The **AVG** function calculates the average (mean) value of a numeric column across the rows returned by a query.

The **SUM** function adds together the values in a numeric column to return the total.

```python
res = cur.execute("""
    SELECT AVG(vote_average) FROM movies
""")

print(res.fetchall())

res = cur.execute("""
    SELECT SUM(budget) FROM movies
""")

print(res.fetchall())
```

## MIN and MAX

The **MIN** function returns the smallest value in a column.

The **MAX** function returns the largest value in a column.

```python
res = cur.execute("""
    SELECT MAX(vote_count), title FROM movies
""")

print(res.fetchall())

res = cur.execute("""
    SELECT MIN(vote_average), title FROM movies
""")

print(res.fetchall())
```

## Aliases

An **alias** is a temporary name given to a column or table in a query. Aliases make query results easier to read and allow you to reference columns or tables with shorter or clearer names.

Aliases are created using the AS keyword and only apply to the result of the query.

```python
res = cur.execute("""
    SELECT COUNT(*) AS total_movies, 
    AVG(budget) AS average_budget
    FROM movies
""")

print([col[0] for col in res.description])

print(res.fetchall())
```

# Filtering Results

## WHERE

The **WHERE** clause is used to filter rows in a query based on specified conditions.

Only rows that satisfy the condition in the WHERE clause are included in the query results.

Common operators include comparison operators (such as =, !=, <, >, <=, >=) and logical operators (such as AND, OR, and NOT) that combine or modify conditions.

```python
res = cur.execute("""
    SELECT title, revenue FROM movies 
    WHERE revenue > 1000000000
    LIMIT 5
""")

res.fetchall()
```

## LIKE

The **LIKE** operator is used in a WHERE clause to search for patterns in text values.

It often uses wildcard characters such as `%`, which represents any sequence of characters, and `_`, which represents a single character.

```python
res = cur.execute("""
    SELECT title, tagline FROM movies
    WHERE tagline LIKE "%adventure%"
    LIMIT 5
""")

res.fetchall()
```

## IN

The **IN** operator is used in a WHERE clause to check whether a value matches any value in a specified list.

It allows you to test a column against multiple possible values without writing multiple conditions.

```python
res = cur.execute("""
    SELECT title, director_id FROM movies
    WHERE director_id IN (4762, 4763)
    LIMIT 5
""")

res.fetchall()
```

## AND and OR

The **AND** operator combines multiple conditions and returns rows only when all conditions are true.

The **OR** operator combines multiple conditions and returns rows when at least one condition is true.

```python
res = cur.execute("""
    SELECT title, revenue, popularity FROM movies
    WHERE (revenue > 500000000 AND popularity < 10)
    OR (revenue < 100000000 AND popularity > 100)
""")

res.fetchall()
```

## NULL

A **NULL** value represents missing or unknown data in a table.

Because NULL is not equal to any value, it must be checked using the IS NULL or IS NOT NULL conditions in a WHERE clause.

```python
res = cur.execute("""
    SELECT title FROM movies
    WHERE tagline IS NULL
    LIMIT 5
""")

res.fetchall()
```

# Grouping, Summarizing, and Sorting

## GROUP BY

The **GROUP BY** clause is used to group rows that have the same values in specified columns.

It is commonly used with aggregate functions to calculate summaries for each group.

```python
res = cur.execute("""
    SELECT director_id, SUM(revenue) AS total_revenue
    FROM movies
    GROUP BY director_id
    LIMIT 5
""")

res.fetchall()
```

## HAVING

The **HAVING** clause is used to filter groups created by a GROUP BY clause.

It is typically used with aggregate functions to restrict which grouped results are included in the final output.

```python
res = cur.execute("""
    SELECT director_id, SUM(revenue) AS total_revenue
    FROM movies
    GROUP BY director_id
    HAVING total_revenue > 5000000000
    LIMIT 5
""")

res.fetchall()
```

## ORDER BY

The **ORDER BY** clause is used to sort the results of a query based on one or more columns.

Results can be sorted in ascending order (ASC) or descending order (DESC).

```python
res = cur.execute("""
    SELECT director_id, SUM(revenue) AS total_revenue
    FROM movies
    GROUP BY director_id
    ORDER BY total_revenue DESC
    LIMIT 5
""")

res.fetchall()
```

# Joining Tables

## LEFT JOIN

A **LEFT JOIN** combines rows from two tables by matching values in related columns.

All rows from the left table are returned, and matching rows from the right table are included when they exist. If there is no match, the result contains NULL values for the columns from the right table.

```python
res = cur.execute("""
    SELECT directors.id, directors.name, movies.title, movies.revenue FROM directors
    LEFT JOIN movies
    ON directors.id = movies.director_id
    LIMIT 5
""")

pprint(res.fetchall())
```

## INNER JOIN

An **INNER JOIN** combines rows from two tables by matching values in related columns.

Only rows with matching values in both tables are included in the result.

## OUTER JOIN

An **OUTER JOIN** returns rows that have matching values in both tables as well as rows that do not have matches.

When no match exists, the result includes NULL values for the columns from the table without a matching row.

## Full Query Example

```python
res = cur.execute("""
    SELECT directors.id, 
        directors.name, 
        SUM(movies.revenue) AS total_revenue
    FROM directors
    LEFT JOIN movies
    ON directors.id = movies.director_id
    WHERE movies.vote_average > 7
        AND movies.vote_count > 4000
        AND movies.tagline LIKE "%adventure%"
    GROUP BY directors.id
    HAVING total_revenue > 500000000
    ORDER BY total_revenue DESC
""")

res.fetchall()
```
