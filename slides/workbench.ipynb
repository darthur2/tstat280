# Python & Data Science Environment Setup Guide

We will be learning up to set up your data science workbench. Setting up your workbench includes:

- Downloading and installing the Python interpreter
- Familiarizing yourself with the terminal
- Learning about Git and how to use Git via GitHub Desktop
- Learning about virtual environments, why they are important, and how to create one
- Familiarizing yourself with and installing several common Python libraries used by data scientists
- Learning about Jupyter Notebooks and JupyterLab and how to run JupyterLab on your own computer

# What is a Python Interpreter

A Python interpreter is a program that reads and executes Python code. When you write Python code, the interpreter runs it line by line.

## Download and Install Python

Go to:

https://www.python.org/downloads/

- Download Python 3.x
- On Windows: Check "Add Python to PATH"
- On Mac: Use the official installer package

# Checking Python Installation

To make sure we've installed Python correctly, open your terminal (either Powershell on Windows or Terminal on Mac) and try one of the following:

- python --version
- python3 --version
- py --version

This should return the version of Python you downloaded (something like Python 3.12.x)

# What is a Terminal?

A terminal is a text-based interface that allows you to interact with your computer using commands.

- Windows -> PowerShell
- Mac -> Terminal

You use the terminal to:

- Navigate folders
- Create directories
- Install packages
- Run Python, Jupyter Notebooks, or JupyterLab (more on that later)

# Basic Terminal Commands

- ls: Lists the files in the current directory
- cd folder_name: Changes the directory to the directory named "folder_name"
- cd ..: Goes up one directory
- mkdir folder_name: Creates a new directory in the current directory named "folder_name"
- pwd: Prints the current working directory

# What is Git

Git is a version control system. It allows you to:

- Track changes in code
- Save snapshots of your work
- Revert to earlier versions
- Collaborate with others

# Creating a GitHub Account

GitHub is a website that stores Git repositories online.

# Using Git via GitHub Desktop

We will use GitHub Desktop for simplicity. Download it here:

https://desktop.github.com/

# Basic Git Workflow (Using GitHub Desktop)

Cloning means downloading a repository from GitHub to your computer. After copying and pasting the the link to the repository, do the following in GitHub Desktop:

- File -> Clone Repository
- Select URL
- Input URL for repository
- Choose local folder where repository will be stored
- Click clone
