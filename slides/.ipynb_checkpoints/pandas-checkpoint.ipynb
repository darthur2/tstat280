## Introduction to Pandas

**Pandas** is one of the most widely used Python libraries for **data analysis and data manipulation**. It provides powerful tools that make it easier to work with structured data such as tables, spreadsheets, or CSV files.

---

### Why Pandas?

Real-world datasets are often:
- Large
- Messy
- Stored in files like CSV or Excel
- Organized in rows and columns

Pandas helps us **load, clean, explore, and analyze this data efficiently**.

---

Most of the time, we will load data into a **DataFrame**.

## Inspecting Pandas DataFrame

Once a dataset has been loaded into a pandas **DataFrame**, several built-in attributes and methods can help you quickly understand its structure and contents.

| Method / Attribute | Description |
|---|---|
| `head()` | Displays the first few rows of the DataFrame (5 by default). Useful for quickly previewing the dataset. |
| `tail()` | Displays the last few rows of the DataFrame (5 by default). Helpful for checking the end of the dataset. |
| `shape` | Returns the dimensions of the DataFrame as a tuple `(rows, columns)`. |
| `columns` | Returns the names of all columns in the DataFrame. |
| `info()` | Provides a summary of the DataFrame including column names, data types, number of non-null values, and memory usage. |
| `describe()` | Generates summary statistics (count, mean, standard deviation, min, quartiles, max) for numerical columns. |

## Selecting and Creating Columns in Pandas

In pandas, columns in a DataFrame can be **selected, modified, or created** using simple indexing operations. This allows you to focus on specific variables or generate new ones from existing data.

You can select and create columns using **square bracket notation**.

```python
df["column_name"] # selects a single column
df[["column_name_1", "column_name_2"]] # selects multiple columns
df["new_column_name"] = new_column_values # creates new column with values in new_column_values 
```

````python
stress["Age"].head()
````

````python
stress[["Daily_Phone_Hours", "Social_Media_Hours"]].head()
````

````python
stress["Sleep_Stress_Ratio"] = stress["Sleep_Hours"]/stress["Stress_Level"]
stress[["Sleep_Hours", "Stress_Level", "Sleep_Stress_Ratio"]].head()
````

## Filtering Rows in a Pandas DataFrame

Often we only want to work with **rows that meet certain conditions**.  
In pandas, this is done using **Boolean filtering** (also called Boolean indexing).

We create a **condition** that evaluates to `True` or `False` for each row.

````python
stress[stress["App_Usage_Count"] > 10].head()
````

````python
stress[(stress["Caffeine_Intake_Cups"] > 3) & (stress["Age"] < 30)].head()
````

## Selecting Data in Pandas with `loc` and `iloc`

In pandas, the two most common ways to select rows and columns from a DataFrame are **`loc`** and **`iloc`**. Both are used for indexing, but they work in different ways:

- **`loc` → label-based indexing**
- **`iloc` → integer position-based indexing**

Understanding the difference is essential for reliable data manipulation.

````python
stress.index = stress["User_ID"]
stress.loc[["U1", "U3", "U5"]]
stress.iloc[[0, 2, 4]]
stress.loc[["U1", "U3", "U5"], ["Age", "Gender"]]
stress.iloc[[0, 2, 4], [1, 2]]
````

## Grouping and Aggregating Data with `groupby`

In many data analysis tasks, we want to **summarize data by categories**.  
For example, we might want to compute:

- The average GPA by major
- The total sales by region
- The number of students in each class year

In pandas, this is done using the **`groupby()`** method combined with **aggregation functions**.

````python
stress.groupby("Gender")["Social_Media_Hours"].mean()
````

````python
stress.groupby("Occupation")[["Stress_Level", "Sleep_Hours"]].agg(["mean", "median"])
````

## Sorting Data in a Pandas DataFrame

Sometimes we want to **reorder rows** in a dataset based on the values in one or more columns.  
In pandas, this is done using the **`sort_values()`** method.

````python
stress.sort_values("Sleep_Hours", ascending=False).head()
````

````python
stress.sort_values(["Occupation", "Sleep_Hours"], ascending=True).head()
````

## Lambda Functions in Python

A **lambda function** is a small, anonymous function defined in a single line.  
Lambda functions are useful when you need a **short function for a simple operation**, especially when working with pandas.

Python function:

```python
def square(x):
    return x**2
```

Corresponding Lambda function:

```python
lambda x: x**2
```

````python
stress["Age"].apply(lambda x: x**2).head()
````

````python
stress.apply(lambda row: row["Social_Media_Hours"] + row["Weekend_Screen_Time_Hours"], axis = 1).head()
````

````python
stress["Burnout_Risk"] = stress.apply(
    lambda row: "High"
    if row["Stress_Level"] > 7 and row["Sleep_Hours"] < 6
    else "Medium"
    if row["Stress_Level"] > 5
    else "Low",
    axis=1
)
````

## The `map()` Method in Pandas

The **`map()`** method is used to **transform the values in a single column (a pandas Series)**. It applies a function or mapping to **each individual value** in that column.

`map()` is commonly used to:
- Convert categorical values to numeric codes
- Replace values using a dictionary
- Apply simple transformations to each element

````python
stress.info()
````

````python
occupation_map = {
    "Professional": "P",
    "Student": "S",
    "Business Owner": "BO",
    "Freelancer": "F"
}

stress["Occupation_Short"] = stress["Occupation"].map(occupation_map)
stress[["Occupation", "Occupation_Short"]].head()
````

## Reading Data from a SQL Database with Pandas

Pandas can also load data directly from a **SQL database** using a SQL query.  
This allows you to retrieve only the data you need instead of loading an entire table.

The main function used for this is **`pd.read_sql()`**.

---

### Basic Syntax

```python
pd.read_sql(query, connection)

````python
import sqlite3

conn = sqlite3.connect("../data/movies.sqlite")

movies = pd.read_sql("""
SELECT * FROM movies
""", conn)

directors = pd.read_sql("""
SELECT * FROM directors
""", conn)

movies.head()
````

````python
directors.head()
````

## Joining Data in Pandas

Often data is stored in **multiple tables**. To analyze it together, we need to **combine datasets based on a shared column**. In pandas, this is done using the **`merge()`** function.

This is similar to **SQL joins**.

---

### Basic Syntax

```python
pd.merge(left_df, right_df, on="column_name")

````python
director_movies = pd.merge(directors, movies, left_on="id", right_on="director_id")
director_movies.head()
````

## Working with Date and Time in Pandas

Many datasets include **date or time information**, such as timestamps, dates of events, or times of activity. Pandas provides powerful tools for converting and working with these values.

---

### Converting to Datetime

Dates are often stored as **strings** when data is first loaded.  
We can convert them to a proper datetime format using **`pd.to_datetime()`**.

```python
df["date"] = pd.to_datetime(df["date"])

````python
director_movies["release_date"] = pd.to_datetime(director_movies["release_date"])
director_movies["release_date"].head()
````

## Working with Dates in Pandas

Once a column has been converted to a **datetime type**, pandas provides convenient tools for extracting parts of the date and filtering data by time.

| Function | Description | Example |
|---|---|---|
| `.dt.year` | Extracts the year from a datetime value | `df["date"].dt.year` |
| `.dt.month` | Extracts the month (1–12) from a datetime value | `df["date"].dt.month` |
| `.dt.month_name()` | Returns the name of the month (e.g., January, February) | `df["date"].dt.month_name()` |
| `.dt.day` | Extracts the day of the month | `df["date"].dt.day` |
| `.dt.day_name()` | Returns the name of the day (e.g., Monday, Tuesday) | `df["date"].dt.day_name()` |
| Date filtering | Filters rows based on a date condition | `df[df["date"] > "2024-01-01"]` |

````python
director_movies["Year"] = director_movies["release_date"].dt.year
director_movies["Month"] = director_movies["release_date"].dt.month_name()
director_movies["Day"] = director_movies["release_date"].dt.day_name()

director_movies[["title", "Year", "Month", "Day"]].head()
````
