```python
import pandas as pd

yt = pd.read_csv("../data/youtube.csv")

yt.head()
```

```python
yt.info()
```

```python
yt["timestamp"] = pd.to_datetime(yt["timestamp"])

yt["category"].unique()
yt["category"] = yt["category"].str.strip().str.lower()
yt["category"].unique()
```

```python
category_dict = {
    "vlog": "vlogs",
    "vlogs": "vlogs",
    "educ": "education",
    "education": "education",
    "gami": "gaming",
    "gaming": "gaming",
    "musi": "music",
    "music": "music",
    "tech": "tech reviews",
    "tech reviews": "tech reviews",
    "codi": "coding",
    "coding": "coding"
}

yt["category"] = yt["category"].map(category_dict)

yt["category"].unique()
```

```python
yt["language"].unique()

yt["language"] = yt["language"].str.split(r"[,;]").map(
    lambda x: x[0]
)
```

```python
yt["region"].unique()

yt["region"] = yt["region"].str.upper()

yt["region"].unique()
```

```python
yt["duration_sec"] = yt["duration_sec"].str.split().map(
    lambda x: float(x[0]) if len(x) == 1
    else float(x[0])*60 if x[1] == "min"
    else float(x[0])
)
```

```python
yt["views"] = yt["views"].fillna(yt["views"].median())
```

```python
yt[["comments", "shares"]] = yt["comments_shares"].str.split(r";", expand=True).astype(int)

yt.drop("comments_shares", axis=1, inplace=True)
```

```python
yt["sentiment_score"] = yt["sentiment_score"].map(
    lambda x: x/100 if x < -1 or x > 1
    else x
)

yt.info()
```
