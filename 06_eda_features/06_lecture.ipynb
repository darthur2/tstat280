## Count Plots (Bar Chart)

Plot the number of observations in each category as a bar chart.

```python

```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **x / y** → categorical variable to count (choose one)  
- **hue** → optional second category for grouping

## Histograms

Plot the distribution of a numerical variable using a histogram (optionally with grouping or density curves).

```python
seaborn.histplot(data=df, x="value", bins=20, hue="group", stat="count")
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **x / y** → numerical variable to plot (choose one)  
- **bins** → number of bins (bars) in the histogram  
- **hue** → optional grouping variable  
- **stat** → what to display (e.g., "count", "density")

## Histogram (Example)

````python
sns.histplot(data=sm, x="Avg_Daily_Usage_Hours")
````

## Grouped Histograms (Example)

````python
sns.histplot(data=sm, x="Avg_Daily_Usage_Hours", hue="Affects_Academic_Performance")
````

## Boxplots

Draw a box-and-whisker plot to show the distribution of a numerical variable across categories.

```python
seaborn.boxplot(data=df, x="category", y="value", hue="group")
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **x / y** → categorical and numerical variables (one for each axis)  
- **hue** → optional grouping variable for nested boxes  

## Boxplot (Example)

````python
sns.boxplot(data=nyc, x="cost_of_the_order")
````

## Grouped Boxplots (Example)

````python
sns.boxplot(data=dmd, x="carat", hue="color")
````

## Density Plots

Plot the estimated probability density function of a continuous variable using kernel density estimation.

```python
seaborn.kdeplot(data=df, x="value", hue="group", common_norm=True)
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **x / y** → numerical variable to estimate density for (choose one or both for 2D)  
- **hue** → optional grouping variable for multiple distributions
- **common_norm** → boolean for if area under the curve should be 1 or not

## Density Plot (Example)



````python
sns.kdeplot(data=dmd, x="price")
````

## Grouped Density Plots (Example)

````python
sns.kdeplot(data=sm, x="Avg_Daily_Usage_Hours", hue="Academic_Level", common_norm=False)
````

## Line Chart (Time Series)

Plot a line chart showing the relationship between two numerical variables (often over time).

```python
seaborn.lineplot(data=df, x="time", y="value", hue="group", style="group")
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **x / y** → numerical variables for the axes  
- **hue** → optional grouping variable for multiple lines  
- **style** → vary line style by a categorical variable  

## Line Chart (Example)

````python
df = pd.DataFrame({"date": ["01-01-2016", "01-01-2017", "01-01-2018", "01-01-2019"],
                   "revenue": [10000, 22500, 40000, 45000]})

sns.lineplot(data=df, x="date", y="revenue")
````

## Linear Model Plot

Plot a scatter chart with a fitted linear regression line.

```python
seaborn.lmplot(data=df, x="x_value", y="y_value", hue="group_variable", ci=None)
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **x / y** → numerical variables for the axes  
- **hue** → optional grouping variable for separate regression lines  
- **ci** → size of confidence interval around regression line

## Linear Model Plot (Example)

````python
sns.lmplot(data=dmd, x="carat", y="price", hue="cut", ci=None)
````

## Pair Plot

Plot pairwise relationships between numerical variables in a dataset.

```python
seaborn.pairplot(data=df, hue="categorical_variable", vars=["var1", ..., "varK"])
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **hue** → optional categorical variable for color grouping  
- **vars** → subset of variables to plot  
- **diag_kind** → type of plot on diagonal (e.g., "hist", "kde")

````python
sns.pairplot(sm, vars=["Avg_Daily_Usage_Hours", "Sleep_Hours_Per_Night", 
                       "Mental_Health_Score", "Conflicts_Over_Social_Media",
                       "Addicted_Score"])
````

## Heatmap

Plot a matrix as a color-encoded grid.

```python
seaborn.heatmap(data=matrix)
```

**Key parameters:**
- **data** → 2D dataset (e.g., correlation matrix or pivot table)  
- **cmap** → color map for values  
- **annot** → show values in cells (True/False)  
- **fmt** → formatting string for annotations  
- **linewidths** → width of lines between cells

## Correlation Heatmap

````python
corr_matrix = df[["Avg_Daily_Usage_Hours", "Sleep_Hours_Per_Night", 
                       "Mental_Health_Score", "Conflicts_Over_Social_Media",
                       "Addicted_Score"]].corr()

sns.heatmap(corr_matrix, annot=True)
````

## Contingency Table Heatmap

````python
acad_platform = pd.crosstab(sm["Academic_Level"], sm["Most_Used_Platform"])

sns.heatmap(acad_platform, annot=True)
````

## Pandas Dummy Variables

Convert categorical variables into one-hot encoded (dummy) variables.

```python
pandas.get_dummies(data=df, columns=["category"])
```

**Key parameters:**
- **data** → dataset (e.g., DataFrame)  
- **columns** → list of categorical columns to encode  
- **drop_first** → drop first level to avoid multicollinearity (True/False)  
- **dtype** → data type of the new columns

## One-Hot Encoding (Example)

````python
pd.get_dummies(data=nyc, columns=["cuisine_type"], dtype="int")
````
