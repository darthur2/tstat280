## The Data Science Life Cycle

### Amazon Laptop Dataset

Jordan Blake has been reviewing the company’s growing dataset of laptop listings collected from several online retailers. The dataset contains information such as product titles, prices, discounts, customer ratings, review counts, and technical details extracted from product pages. Jordan wants the data scientist to use this large dataset to support better decision-making about how laptops should be priced and marketed on the platform.

One of Jordan’s goals is to create a system that can estimate what price a laptop should sell for based on its specifications and listing information—things like processor type, RAM, storage, brand, and display size that appear in the product titles and technical details. Jordan also wants a tool that can automatically determine whether a laptop listing will likely receive strong customer ratings based on its specifications, price, and early engagement signals such as review counts. In addition, Jordan is curious whether the laptops in the dataset naturally fall into meaningful groups. By identifying clusters of similar devices—such as budget laptops, gaming laptops, and premium ultrabooks—the company could better understand the structure of the market and improve how products are categorized and recommended to customers.

The dataset was assembled by Daniel Kim, who built automated data pipelines that gather laptop listings from e-commerce websites through web scraping and API requests. His scripts collect product titles, pricing information, discounts, ratings, review counts, and technical specification fields, then store everything in a structured dataset so it can be analyzed using tools like Python and Pandas.

Maya Rodriguez has already begun exploring the dataset and can help by performing exploratory analysis. She can calculate summary statistics, create visualizations of price distributions and ratings, and investigate relationships between features such as brand, price, and customer feedback. Her exploratory work will help identify patterns, detect outliers, and guide which variables are most useful when building models.

Liam Chen will help once the models are ready to be used in practice. He can take the models developed using tools such as Scikit-Learn and integrate them into the company’s data systems so predictions can be generated automatically when new laptop listings appear. Liam can also help build processes that regularly retrain the models as more data is collected.

Priya Nair has extensive experience in the laptop and consumer electronics market. She understands how different specifications influence consumer purchasing decisions and how brands position their devices. Priya can help interpret patterns discovered in the analysis, confirm whether the groups of laptops identified by the models make sense from a market perspective, and suggest additional features—such as processor generation or GPU class—that may improve the models.
