### Creating a New GitHub Repository

#### Part 1

Create a new GitHub repository in your GitHub account titled `tstat280_practice`

#### Part 2

Clone this repository to your personal machine.

#### Part 3

Open the terminal on your machine and navigate to the cloned project.

#### Part 4

Run the command to start running the Jupyter Lab locally.

#### Part 5

Create a new notebook titled `practice_notebook` and add a markdown cell to the notebook. Fill the notebook with the following text:
