# The Data Science Life Cycle

## What is Data Science?

**Data science** is the field of extracting knowledge and insights from data using statistics, computing, and domain expertise.

**Examples of core activities:**

* Collect and manage data
* Clean and prepare data
* Explore and visualize data
* Build statistical or machine learning models
* Communicate insights to support decisions


## Why Data Science?

Organizations use **data science** to turn data into better decisions.

**Why it matters:**
- Make **data-driven decisions**
- **Discover patterns** in large datasets
- **Predict future outcomes**
- **Improve efficiency** and reduce costs
- Gain **competitive advantage**

**Examples:**
- Product recommendations  
- Fraud detection  
- Sales forecasting  
- Operations optimization

## The Data Science Workflow

There is **no single authoritative definition** of the data science workflow.  
Different organizations use different processes, but most workflows include similar stages.

**Common steps:**
1. Define the **problem or question**
2. **Collect and acquire data**
3. **Clean and prepare** the data
4. **Explore and analyze** the data
5. **Build models** or apply statistical methods
6. **Evaluate results**
7. **Communicate insights** and support decisions

## Problem Definition

The **first step** in the data science workflow is clearly defining the problem.

**Key questions:**
- What decision needs to be made?
- What question are we trying to answer?
- Who are the stakeholders?
- What does **success** look like?

**Why it matters:**
- Aligns the project with **business goals**
- Determines what **data** is needed
- Guides the **analysis and modeling approach**

## Supervised Learning

**Supervised learning** uses data with **known outcomes (labels)** to train models.

**Goal:**
- Learn a relationship between **inputs (features)** and **outputs (labels)**
- Use that relationship to **predict outcomes for new data**

**Examples:**
- **Regression:** Predicting a student's **final exam score** based on study hours and attendance  
- **Classification:** Classifying **emails as spam or not spam**

## Unsupervised Learning

**Unsupervised learning** works with data that **does not have labels**.

**Goal:**
- Discover **patterns, structure, or groupings** in the data

**Examples:**
- **Customer segmentation** based on purchasing behavior  
- **Grouping news articles** by similar topics

## Collect and Acquire Data

Once the problem is defined, the next step is gathering the data needed for analysis.

**Common data sources:**
- Databases and data warehouses  
- APIs and web services  
- Surveys and experiments  
- Sensors, logs, or transaction records  
- Public or open datasets

**Key considerations:**
- Is the data **relevant** to the problem?
- Is the data **complete and reliable**?
- Do we have permission to **access and use** the data?

## Clean and Prepare Data

Raw data is often **messy or incomplete**, so it must be cleaned and prepared before analysis.

**Common tasks:**
- Handle **missing values**
- Fix **errors or inconsistencies**
- Remove **duplicates**
- Convert data into usable **formats**
- Create new **features or variables**

**Why it matters:**
- Improves **data quality**
- Makes data **ready for analysis and modeling**

## Explore and Analyze Data

Before building models, data scientists explore the data to understand its structure and patterns.

**Common activities:**
- Calculate summary statistics
- Visualize data with charts and plots
- Identify patterns, trends, or anomalies
- Examine relationships between variables

**Goals:**
- Develop intuition about the dataset
- Identify potential issues or outliers
- Inform the choice of modeling approaches

## Build Models

In this stage, data scientists apply statistical or machine learning methods to learn patterns from the data.

**Common activities:**
- Select an appropriate modeling approach
- Train models using historical data
- Tune model parameters
- Compare multiple models

**Examples:**
- Regression for predicting numeric values
- Classification for predicting categories
- Clustering to find natural groupings in data

## Evaluate Models and Results

After building models, the next step is to determine how well they perform.

**Common activities:**
- Test models on **unseen data**
- Calculate **performance metrics**
- Compare results across multiple models
- Check for **overfitting or bias**

**Examples of metrics:**
- Accuracy
- Mean squared error

**Goal:**
- Ensure the model provides **reliable and useful results**.

## Communicate Results and Deploy Models

The final step is turning analysis into **actionable outcomes**.

**Communicating results:**
- Present findings with **visualizations and reports**
- Explain insights to **stakeholders**
- Provide **recommendations** for decisions

**Model deployment:**
- Integrate models into **applications or systems**
- Automate predictions or decision support
- Monitor model **performance over time**

**Goal:**
- Ensure insights and models create **real-world impact**.

## Roles in Data Science

Data science projects often involve **multiple roles working together**.

**Common roles:**
- **Domain Experts / Stakeholders** – define the problem and business goals
- **Data Engineers** – collect, store, and manage data
- **Data Scientists / Statisticians** – analyze data and build models
- **Data Analysts** – explore data and communicate insights
- **Machine Learning Engineers** – deploy and maintain models in production

## Domain Experts / Stakeholders

**Domain experts and stakeholders** help define the problem and ensure the work is relevant to the organization.

**Responsibilities:**
- Define the **business problem or question**
- Provide **context and domain knowledge**
- Help determine what **success looks like**
- Use insights to **inform decisions**

**Examples:**
- Marketing managers  
- Healthcare professionals  
- Financial analysts  
- Product managers

## Data Engineers

**Data engineers** build and maintain the systems that collect, store, and process data.

**Responsibilities:**
- Build **data pipelines** to collect and move data
- Manage **databases and data warehouses**
- Ensure data is **reliable, accessible, and scalable**
- Prepare data for **analysis and modeling**

**Tools often used:**
- SQL
- Distributed data systems
- Data pipelines and ETL workflows

## Data Scientists / Statisticians

**Data scientists and statisticians** analyze data to extract insights and build predictive models.

**Responsibilities:**
- Explore and analyze datasets
- Apply **statistical methods** and **machine learning**
- Build and evaluate predictive models
- Interpret results and generate insights

**Common tools:**
- Python or R
- Statistical modeling
- Machine learning methods

## Data Analysts

**Data analysts** focus on exploring data and communicating insights that support decision-making.

**Responsibilities:**
- Query and analyze data
- Create **reports and dashboards**
- Visualize data to highlight trends
- Communicate insights to stakeholders

**Common tools:**
- SQL
- Excel or spreadsheets
- Data visualization tools (e.g., Tableau, Power BI)

## Machine Learning Engineers

**Machine learning engineers** focus on deploying and maintaining models in real-world systems.

**Responsibilities**
- Deploy models into **applications or production systems**
- Build **APIs and services** for model predictions
- Monitor **model performance over time**
- Retrain and update models as new data arrives

**Common tools**
- Python
- Cloud platforms
- Model serving and monitoring tools

# The Data Science Workbench

## Common Tools in the Data Science Workbench

Data scientists use a variety of tools to **write code, manage data, collaborate, and communicate results**.

**Common categories of tools**
- **Programming languages** – Python, R
- **Development environments** – JupyterLab, VS Code
- **Version control** – Git and GitHub
- **Command line tools** – Terminal or PowerShell
- **Data science libraries** – NumPy, Pandas, Matplotlib, Scikit-Learn

These tools together form the **data science workbench** used to develop and run data science projects.

## Python and the Python Interpreter

**Python** is one of the most widely used programming languages in data science.

**Why Python is popular:**
- Easy to **learn and read**
- Large ecosystem of **data science libraries**
- Widely used in **industry and research**

**Python Interpreter:**
- A program that **reads and executes Python code**
- Runs code **line by line**
- Allows you to run Python scripts or interactively execute commands

## Installing Python (Windows)

```bash
py --version
```

## Git

**Git** is a **version control system** used to track changes in code and collaborate with others.

**What Git does:**
- Tracks changes to files over time
- Allows you to **save versions** of your work
- Helps teams **collaborate on the same project**

**Key concepts:**
- **Repository (repo):** a project folder tracked by Git
- **Commit:** a saved snapshot of changes
- **Push:** upload changes to a remote repository
- **Pull:** download updates from a remote repository

## GitHub

**GitHub** is an online platform for **hosting and collaborating on Git repositories**.

**What GitHub provides:**
- A place to **store and share repositories**
- Tools for **collaboration and code review**
- A way to **track issues and project progress**

**Common actions:**
- **Create repositories** to store projects
- **Fork repositories** to copy someone else's project
- **Clone repositories** to work on them locally
- **Push changes** from your computer to GitHub

## Creating a GitHub Account

**1. Go to GitHub**
- https://github.com → Click **Sign up**

**2. Create your account**
- Enter **email, username, and password**

**3. Verify your account**
- Complete the **verification step**

**4. Finish setup**
- Follow the prompts to complete your profile

## What is a Repository?

A **repository (repo)** is a folder that stores a project and its history using Git.

**What a repository contains**
- Project **files and code**
- **Version history** of changes
- Documentation (often a **README** file)

**Why repositories are useful**
- Track changes over time
- Collaborate with others
- Manage and organize projects

## Creating a GitHub Repository

A **repository** is where your project files and history are stored on GitHub.

**Why create a repository**
- Store and organize **project files**
- **Track changes** with Git
- **Share or collaborate** with others

**How to create a repository**
1. Click the **+** icon on GitHub  
2. Select **New repository**  
3. Enter a **repository name**  
4. Choose **Public** or **Private**  
5. Click **Create repository**


## Creating a Repository from a Template

A **template repository** lets you create a **new project that starts with existing files**.

**Why use a template**
- Start with **pre-made project files**
- Keep your work **separate from the original**
- Ensure everyone begins with the **same materials**

**How to create a repository from a template**
1. Go to the **template repository**
2. Click **Use this template**
3. Select **Create a new repository**
4. Enter a **repository name**
5. Click **Create repository**


## Adding Someone With View Access

You can allow someone to **view your repository without editing it**.

**Why give view access**
- Let an instructor **review your work**
- Share your project **without allowing edits**

**How to add view access**
1. Go to your **repository**
2. Click **Settings**
3. Select **Collaborators and teams**
4. Click **Add people**
5. Enter the **GitHub username**
6. Set role to **Read**


## GitHub Desktop

**GitHub Desktop** is a graphical application that makes it easier to work with **Git and GitHub**.

**What it allows you to do**
- Clone repositories to your computer
- Track and commit file changes
- Push updates to GitHub
- Pull updates from a repository

**Why use GitHub Desktop**
- No need to remember Git commands
- Simple interface for beginners
- Easy way to manage repositories

## Installing GitHub Desktop

**1. Download GitHub Desktop**
- Go to: https://desktop.github.com

**2. Install the application**
- Download and run the installer
- Follow the setup instructions

**3. Sign in to GitHub**
- Open **GitHub Desktop**
- Sign in with your **GitHub account**

You can now **clone and manage repositories** from your computer.

## Cloning a Repository (GitHub Desktop)

**Cloning** creates a **local copy of a repository** on your computer.

**Steps**
1. Open **GitHub Desktop**
2. Click **File → Clone Repository**
3. Select the repository from the **GitHub.com** tab
4. Choose the **local folder** where it will be saved
5. Click **Clone**

You can now **edit the project files locally** on your computer.

## Commit and Push

**Commit**
- A **commit** saves a snapshot of your changes in the repository.
- Each commit includes a **message** describing what changed.

**Push**
- **Push** sends your committed changes from your computer to **GitHub**.

**Typical workflow**
1. Make changes to files
2. **Commit** the changes in GitHub Desktop
3. **Push** the commits to GitHub

## What is a Terminal?

A **terminal** is a program that lets you interact with your computer using **text commands**.

**What it is used for**
- Running programs
- Navigating files and folders
- Managing software and environments
- Executing commands for tools like Python or Git

**Examples**
- **Windows:** PowerShell or Command Prompt  
- **Mac/Linux:** Terminal

**Method 1**
1. Click the **Start menu**
2. Type **PowerShell**
3. Click **Windows PowerShell**

**Method 2**
1. Press **Windows Key + X**
2. Select **Windows Terminal** or **PowerShell**

## Opening the Terminal (Mac)

**Method 1**
1. Press **Command + Space** to open Spotlight
2. Type **Terminal**
3. Press **Enter**

**Method 2**
1. Open **Finder**
2. Go to **Applications → Utilities**
3. Click **Terminal**

**ls**
- Lists the **files and folders** in the current directory

**cd**
- Changes the **current directory (folder)**
- Example: `cd Documents`

**pwd**
- Shows the **current directory path**
